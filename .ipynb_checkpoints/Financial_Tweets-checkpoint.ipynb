Dành số 0 cho từ unknown nên shift các index lên + 1
p là pair(p._1, p._2) => phần tử thứ nhất giữ nguyên, phần tử thứ 2 cộng thêm 1

```python
# Convert vocabulary to RDD
vocab_rdd = spark.sparkContext.parallelize(model)

# Use zipWithIndex to add indices (starting from 0)
vocab_with_index = vocab_rdd.zipWithIndex()

# Map to create (word, index + 1) pairs
vocab_with_index_plus_one = vocab_with_index.map(lambda p: (p[0], p[1] + 1))

# Collect the results as a Python list
vocab_dict = vocab_with_index_plus_one.collect()

# Print the result
print(vocab_dict)

# model.vocabulary.zipWithIndex.map(p: p._1, p._2 + 1)
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import ArrayType, IntegerType

# Define your dictionary
vocab_dict = {'language': 1, 'processing': 2, 'natural': 3, 'love': 4}

# Define the UDF function in Python (flatten the list if needed)
def map_to_values(xs):
    # Flatten if xs is a list of lists
    flat_xs = [item for sublist in xs for item in sublist] if isinstance(xs[0], list) else xs
    return [vocab_dict.get(x, 0) for x in flat_xs]  # Using get() to handle missing keys

# Register the UDF with the appropriate return type
map_udf = udf(map_to_values, ArrayType(IntegerType()))

# Assuming your DataFrame has a 'tokens' column that is a list of lists (or just a list of strings)
# Applying the UDF to the DataFrame
df_with_values = df.withColumn("word_values", map_udf("tokens"))

# Show the results
df_with_values.show(truncate=False)
```
